```python
# 9.7.1
# selective copy

# module import 
import os
import shutil

# define function
def selective_copy(str_file_ext, str_folder):
    
    str_abspath = os.path.abspath(str_folder)
    str_dest_dir = os.path.join(os.getcwd(), "prac9_7_1")
    
    if not os.path.exists(str_dest_dir):
        os.makedirs(str_dest_dir)
    
    for str_foldername, lst_subfolders, lst_filenames in os.walk(str_folder):
        print("searching file in directory: {}".format(str_foldername))
        
        if str_foldername == os.path.join(".", "prac9_7_1"):
            break
        
        for str_filename in lst_filenames:
            if str_filename.endswith(str_file_ext):
                str_filepath = os.path.join(str_foldername, str_filename)
                shutil.move(os.path.join(str_foldername, str_filename), str_dest_dir)
                
    print("Finished.")
    return


# test
selective_copy(".txt", ".\\")                
```

```python
# 9.7.2
# search fuge file

# module import 
import os
import shutil

# defin function
def search_huge_file(str_dir, int_thr):
    
    str_dir_path = os.path.abspath(str_dir)
    lst_abspath_return = []
    
    for str_foldername, lst_subfolders, lst_filenames in os.walk(str_dir):
        print("searching file in directory: {}".format(str_foldername))
        
        for str_filename in lst_filenames:
            int_size = os.path.getsize(os.path.join(str_foldername, str_filename))
            
            if int_size > int_thr:
                lst_abspath_return.append(os.path.abspath(str_filename))
                
    if len(lst_abspath_return) != 0:
        print("Huge file is following: ")
        for path in lst_abspath_return:
            print(path)
    else:
        print("No such huge file.")
                
    return


# test
search_huge_file(".\\", int(5*1e3))
```
